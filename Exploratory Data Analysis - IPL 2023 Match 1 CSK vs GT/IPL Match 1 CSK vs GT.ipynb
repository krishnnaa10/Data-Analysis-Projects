# Data Analysis - IPL 2023 Match 1 : CSK vs GT 

## Importing Libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Importing Data

```python
df = pd.read_csv('match_01.csv')
```

```python
df
```

## Cleaning Data

```python
df.info()
```

```python
df.duplicated().sum()
```

## Splitting Data

#### **First inning Data**

```python
df1 = df.iloc[0:121,:]
df1
```

```python
df1 = df1.drop('inningno',axis=1)
df1
```

#### **Second inning Data**

```python
df2 = df.iloc[121:,:]
df2
```

```python
df2 = df2.reset_index()
```

```python
df2 = df2.drop('index',axis=1)
df2 = df2.drop('inningno',axis=1)
df2
```

## Count of ball faced by each batsman

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=df1['batter'], color='gold')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=df2['batter'], color='teal')
plt.xticks(rotation=90)
plt.show()
```

## Count of balls by each bowler

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=df1['bowler'], color='orange')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=df2['bowler'], color='olive')
plt.xticks(rotation=90)
plt.show()
```

## Overall sixes in the match

```python
sixes = df[df['outcome']=='6']
sixes
```

```python
sixes['batter'].value_counts()
```

#### **Count of batmen hit sixes**

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=sixes['batter'], color='purple')
plt.xticks(rotation=90)
plt.show()
```

#### **Count of bowlers concede sixes**

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=sixes['bowler'], color='aqua')
plt.xticks(rotation=90)
plt.show()
```

## Overall fours in the match

```python
fours = df[df['outcome']=='4']
fours
```

```python
fours['batter'].value_counts()
```

#### **Count of batmen hit fours**

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=fours['batter'], color='darkgreen')
plt.xticks(rotation=90)
plt.show()
```

#### **Count of bowlers concede fours**

```python
plt.figure(figsize=(8,5))
ax=sns.countplot(x=fours['bowler'], color='coral')
plt.xticks(rotation=90)
plt.show()
```

## Making Pivot Table

```python
df3 = df1.groupby(['batter', 'outcome'])['outcome'].count()
df3.to_csv('score.csv')
df3 = pd.read_csv('score.csv')
df3 = pd.pivot_table(df3, values=None, index=['batter'], columns=['outcome'], fill_value=0, aggfunc="sum")
df3
```

```python
df4 = df2.groupby(['batter', 'outcome'])['outcome'].count()
df4.to_csv('score.csv')
df4 = pd.read_csv('score.csv')
df4 = pd.pivot_table(df4, values=None, index=['batter'], columns=['outcome'], fill_value=0, aggfunc="sum")
df4
```
